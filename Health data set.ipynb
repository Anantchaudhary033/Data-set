```python
x=df["Calories"].mode()[0]
x
```

```python
# range replace
```

```python
for x in df.index:
    if df.loc[x,"Duration"]>120:
        df.loc[x,"Duration"]=120
print(df.to_string())        
```

```python
# drop
```

```python
for x in df.index:
    if df.loc[x,"Duration"]>120:
        df.drop(x,inplace=True)
print(df.to_string()) 
```
